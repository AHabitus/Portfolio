```python
age_lst = [23, 45, 18, 80, 22, 25, 34, 37, 41, 88, 19, 20, 18, 45, 41, 39, 31, 32, 22, 28, 37]
```

```python
print(round(sum(age_lst) / len(age_lst), 2))
```

```python
sum_age = 0
cnt_age = 0
for i in age_lst:
    if i < 50:
        sum_age += i
        cnt_age += 1
print(round((sum_age / cnt_age), 2))
```

```python
def calc_avg(lst, lim):
    sum_age = 0
    cnt_age = 0
    for i in lst:
        if i < lim:
            sum_age += i
            cnt_age += 1
    return round((sum_age / cnt_age), 2)
```

```python
calc_avg(age_lst, 50)
```

```python
import numpy as np
```

```python
q1 = np.percentile(age_lst, 25)
q3 = np.percentile(age_lst, 75)
iqr = q3-q1
lower_bound = q1-1.5*iqr
upper_bound = q3+1.5*iqr
sum_age = 0
cnt_age = 0
for i in age_lst:
    if (i<upper_bound) and (i>lower_bound):
        sum_age +=i
        cnt_age +=1
avg_age = sum_age/cnt_age
print(f"Новое среднее значение {round(avg_age, 2)}")
```
